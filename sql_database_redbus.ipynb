```python
# csv to dataframe
df_andhra=pd.read_csv(r"C:\Users\kumar\Desktop\scraped_data\busdetailsall\andhra_busdetails.csv")
df_kerala=pd.read_csv(r"C:\Users\kumar\Desktop\scraped_data\busdetailsall\kerala_busdetails.csv")
df_tel=pd.read_csv(r"C:\Users\kumar\Desktop\scraped_data\busdetailsall\telangana_busdetails.csv")
df_har=pd.read_csv(r"C:\Users\kumar\Desktop\scraped_data\busdetailsall\haryana_busdetails.csv")
df_up=pd.read_csv(r"C:\Users\kumar\Desktop\scraped_data\busdetailsall\uttarpradesh_busdetails.csv")
df_raj=pd.read_csv(r"C:\Users\kumar\Desktop\scraped_data\busdetailsall\rajasthan_busdetails.csv")
df_ass=pd.read_csv(r"C:\Users\kumar\Desktop\scraped_data\busdetailsall\assam_busdetails.csv")
df_meg=pd.read_csv(r"C:\Users\kumar\Desktop\scraped_data\busdetailsall\meghalaya_busdetails.csv")
df_gj=pd.read_csv(r"C:\Users\kumar\Desktop\scraped_data\busdetailsall\gujarat_busdetails.csv")
df_pun=pd.read_csv(r"C:\Users\kumar\Desktop\scraped_data\busdetailsall\punjab_busdetails.csv")
df_full1=pd.concat([df_andhra,df_kerala,df_tel,df_har,df_up,df_raj,df_ass,df_meg,df_gj,df_pun],ignore_index=True)
df_full1


```

```python
# Ensure all values in the column are strings
df_full1["Price"] = df_full1["Price"].astype(str)

# Remove "INR" and any other non-numeric characters
df_full1["Price"] = df_full1["Price"].str.replace("INR", "", regex=False).str.strip()

# Convert to numeric, coercing errors to NaN
df_full1["Price"] = pd.to_numeric(df_full1["Price"], errors="coerce")

# Fill NaN values with 0
df_full1["Price"] = df_full1["Price"].fillna(0)
```

```python
df_full1["Ratings"]
```

```python
df_full1["Ratings"] = df_full1["Ratings"].astype(str)
df_full1["Ratings"] = df_full1["Ratings"].str.replace("New", "")
df_full1["Ratings"] = df_full1["Ratings"].str.strip()
df_full1["Ratings"] = df_full1["Ratings"].str.split().str[0]
# Convert to numeric, coercing errors
df_full1["Ratings"] = pd.to_numeric(df_full1["Ratings"], errors='coerce')
df_full1["Ratings"].fillna(0, inplace=True)

print(df_full1["Ratings"])
```
